```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from delta.tables import *
```

```python
df = spark.read.format("csv").option("header","true").load("Files/Player_info.csv")
```

```python
deltaTable = DeltaTable.forPath(spark, 'Tables/dimplayer')
    
dfUpdates = df
    
deltaTable.alias('gold') \
  .merge(
    dfUpdates.alias('updates'),
    'gold.player_name = updates.PlayerName'
  ) \
   .whenMatchedUpdate(set =
    {
        "imageurl": "updates.PlayerImageUrl",
        "born": "updates.Born",
        "country": "updates.Country",
        "batting_style": "updates.BattingStyle",
        "bowling_style": "updates.BowlingStyle",
        "playing_role": "updates.PlayingRole"
    }
  ) \
  .execute()
```

```python
df = spark.read.format("csv").option("header","true").load("Files/team_info.csv")
```

```python
deltaTable = DeltaTable.forPath(spark, 'Tables/dimteam')
    
dfUpdates = df
    
deltaTable.alias('gold') \
  .merge(
    dfUpdates.alias('updates'),
    'gold.name = updates.TeamName'
  ) \
   .whenMatchedUpdate(set =
    {
       "abbr" : "TeamAbbrevation",
       "logourl" : "TeamLogoUrl"   
    }
  ) \
  .execute()
```
