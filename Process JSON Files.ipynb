```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```python
schema = StructType([
    StructField('info', StructType([
        StructField('balls_per_over', LongType(), True),
        StructField('city', StringType(), True),
        StructField('dates', ArrayType(StringType(), True), True),
        StructField('event', StructType([
            StructField('match_number', LongType(), True),
            StructField('name', StringType(), True),
            StructField('stage', StringType(), True)
        ]), True),
        StructField('gender', StringType(), True),
        StructField('match_type', StringType(), True),
        StructField('officials', StructType([
            StructField('match_referees', ArrayType(StringType(), True), True),
            StructField('reserve_umpires', ArrayType(StringType(), True), True),
            StructField('tv_umpires', ArrayType(StringType(), True), True),
            StructField('umpires', ArrayType(StringType(), True), True)
        ]), True),
        StructField('outcome', StructType([
            StructField('by', StructType([
                StructField('runs', LongType(), True),
                StructField('wickets', LongType(), True)
            ]), True),
            StructField('winner', StringType(), True),
            StructField('eliminator', StringType(), True)
        ]), True),
        StructField('overs', LongType(), True),
        StructField('player_of_match', ArrayType(StringType(), True), True),
        StructField('season', StringType(), True),
        StructField('team_type', StringType(), True),
        StructField('teams', ArrayType(StringType(), True), True),
        StructField('toss', StructType([
            StructField('decision', StringType(), True),
            StructField('winner', StringType(), True)
        ]), True),
        StructField('venue', StringType(), True)
    ]), True),
    
    StructField('innings', ArrayType(StructType([
        StructField('overs', ArrayType(StructType([
            StructField('deliveries', ArrayType(StructType([
                StructField('batter', StringType(), True),
                StructField('bowler', StringType(), True),
                StructField('extras', StructType([
                    StructField('byes', LongType(), True),
                    StructField('legbyes', LongType(), True),
                    StructField('noballs', LongType(), True),
                    StructField('penalty', LongType(), True),
                    StructField('wides', LongType(), True)
                ]), True),
                StructField('non_striker', StringType(), True),
                StructField('runs', StructType([
                    StructField('batter', LongType(), True),
                    StructField('extras', LongType(), True),
                    StructField('total', LongType(), True),
                    StructField('non_boundary', BooleanType(), True)
                ]), True),
                StructField('wickets', ArrayType(StructType([
                    StructField('fielders', ArrayType(StructType([
                        StructField('name', StringType(), True),
                        StructField('substitute', BooleanType(), True)
                    ]), True), True),
                    StructField('kind', StringType(), True),
                    StructField('player_out', StringType(), True)
                ]), True), True)
            ]), True), True),
            StructField('over', LongType(), True)
        ]), True), True),
        StructField('powerplays', ArrayType(StructType([
            StructField('from', DoubleType(), True),
            StructField('to', DoubleType(), True),
            StructField('type', StringType(), True)
        ]), True), True),
        StructField('target', StructType([
            StructField('overs', LongType(), True),
            StructField('runs', LongType(), True)
        ]), True),
        StructField('team', StringType(), True)
    ]), True), True),

    StructField('meta', StructType([
        StructField('created', StringType(), True),
        StructField('data_version', StringType(), True),
        StructField('revision', LongType(), True)
    ]), True)
])
```

```python
file_id = '1082591'
file_path = 'Files/Raw JSON/'+file_id+'.json'
```

```python
df = spark.read.option("multiline", "true").schema(schema).json(file_path)
```

```python
#df.printSchema()
```

```python
df_info = df.select('info.*') 
```

```python
df_info.printSchema()
```

```python
display(df_info.select(col('balls_per_over')
                , col('city')
                , col('event.match_number')
                , col('event.name')
                , col('gender')
                , col('match_type')
                , col('event.stage')
                , col('overs')
                , col('player_of_match')[0].alias('player_of_match')
                , col('season')
                , col('team_type')
                , col('toss.decision')
                , col('toss.winner').alias('toss_winner')
                , col('outcome.winner').alias('winner')
                , col('outcome.by.runs').alias('win_by_runs')
                , col('outcome.by.wickets').alias('win_by_wickets')
                , col('venue')
)
)
```

```python
#df_info.printSchema()
```

```python
df_innings = df.select(explode('innings').alias('innings_expanded'))
```

```python
#df_innings.select('innings_expanded.*').show()
```

```python
df_innings = df_innings.withColumn('innings', when(col('innings_expanded.target').isNull(), 1).otherwise(2))
```

```python
#df_innings.printSchema()
```

```python
df_overs = df_innings.select('innings', col('innings_expanded.team').alias('team'), explode('innings_expanded.overs').alias('over'))
```

```python
df_overs = df_overs.select('innings', 'team', 'over.over', explode('over.deliveries').alias('deliveries'))
```

```python
#df_overs.printSchema()
```

```python
df_overs = df_overs.select('innings', 'team', 'over', 'deliveries.*')
```

```python
df_overs = df_overs.withColumn('batter_runs', col('runs.batter'))\
                    .withColumn('extra_runs', col('runs.extras'))\
                    .withColumn('total_runs', col('runs.total'))\
                    .withColumn('non_boundary', col('runs.non_boundary'))\
                    .withColumn('byes', col('extras.byes'))\
                    .withColumn('legbyes', col('extras.byes'))\
                    .withColumn('noballs', col('extras.noballs'))\
                    .withColumn('penalty', col('extras.penalty'))\
                    .withColumn('wides', col('extras.wides'))\
                    .withColumn('wicket', explode_outer('wickets'))\
                    .withColumn('fielder1',col('wicket.fielders')[0])\
                    .withColumn('fielder1', col('fielder1.name'))\
                    .withColumn('fielder2',col('wicket.fielders')[1])\
                    .withColumn('fielder2', col('fielder2.name'))\
                    .withColumn('player_out', col('wicket.player_out'))\
                    .withColumn('wicket', col('wicket.kind'))
```

```python
display(df_overs.select('innings'
                , 'team'
                , 'over'
                , 'batter'
                , 'bowler'
                , 'non_striker'
                , 'batter_runs'
                , 'extra_runs'
                , 'total_runs'
                , 'non_boundary'
                , 'byes'
                , 'legbyes'
                , 'noballs'
                , 'penalty'
                , 'wides'
                , 'wicket'
                , 'player_out'
                , 'fielder1'
                , 'fielder2'
))
```
